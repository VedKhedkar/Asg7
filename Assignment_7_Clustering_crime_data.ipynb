```python
##Clustering-using_DBSCAN

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Import Dataset
crime=pd.read_csv('crime_data.csv')
crime
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

```python
# Normalize heterogenous numerical data using standard scalar fit transform to dataset
crime_norm=StandardScaler().fit_transform(crime)
crime_norm
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime_norm)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Adding clusters to dataset
crime['clusters']=dbscan.labels_
crime
```

```python
crime.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime['clusters'],crime['UrbanPop'], c=dbscan.labels_) 
```

# Clustering-using_K-Means

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
crime=pd.read_csv('crime_data.csv')
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
# Normalize heterogenous numerical data by using Standard Scaler
crime_norm=StandardScaler().fit_transform(crime)
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(crime_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Build Cluster algorithm using K=4

```python
# Cluster algorithm using K=4
clusters4=KMeans(4,random_state=30).fit(crime_norm)
clusters4
```

```python
clusters4.labels_
```

```python
# Assign clusters to the data set
crime4=crime.copy()
crime4['clusters4id']=clusters4.labels_
crime4
```

```python
clusters4.cluster_centers_
```

```python
# Group data by Clusters (K=4)
crime4.groupby('clusters4id').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime4['clusters4id'],crime4['UrbanPop'], c=clusters4.labels_)
```

Build Cluster algorithm using K=5

```python
# Cluster algorithm using K=5
clusters5=KMeans(5,random_state=30).fit(crime_norm)
clusters5
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime5['clusters5id'],crime5['UrbanPop'], c=clusters5.labels_)
```

# Clustering-using_Hierarchical_clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
crime=pd.read_csv('crime_data.csv')
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
# Normalize heterogenous numerical data 
crime_norm=pd.DataFrame(normalize(crime),columns=crime.columns)
crime_norm
```

```python
# Create Dendrograms
plt.figure(figsize=(10, 7))  
dendograms=sch.dendrogram(sch.linkage(crime_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(crime_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
crime['clustersid']=hclusters.labels_
crime
```

```python
crime.groupby('clustersid').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime['clustersid'],crime['UrbanPop'], c=hclusters.labels_)
```
